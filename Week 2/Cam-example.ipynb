```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
datain=np.loadtxt('data\\Killingley2022_throat.csv',delimiter=',')
log10FFA=datain[:,0]
log10qPCR=datain[:,1]
print(log10FFA,log10qPCR)
```

```python
linreg_out=linregress(log10qPCR,log10FFA)
print(linreg_out)
beta=linreg_out.slope
intercept=linreg_out.intercept
log10qPCR_plot=np.linspace(7.0,10.0,100)


plt.plot(log10qPCR,log10FFA,lw=0,marker='o',ms=8,c='g',label='Killingley data')
plt.plot(log10qPCR_plot,intercept+beta*log10qPCR_plot,lw=3,c='b',label='fit')


plt.legend(fontsize=20)
plt.xlabel('log$_{10}$ qPCR',fontsize=18)
plt.ylabel('log$_{10}$ FFA',fontsize=20)
plt.xticks(fontsize=22)
plt.yticks(fontsize=20)
#plt.xlim([1.0,800])
plt.tight_layout()
#plt.savefig('killingley.png')
plt.show()
```

Residuals are by definition just data points minus the predicted values from the fit

Residuals should be close somehow to set of random numbers drawn from a Gaussian distribution of zero mean and some width (width can be estimated).

```python
residuals=log10FFA-intercept-beta*log10qPCR
print('residuals ',residuals)
plt.plot(log10qPCR,residuals,lw=0,marker='o',ms=8,c='g',label='residuals')


plt.legend(fontsize=20)
plt.xlabel('log$_{10}$ qPCR',fontsize=18)
plt.ylabel('residuals',fontsize=20)
plt.xticks(fontsize=22)
plt.yticks(fontsize=20)
#plt.xlim([1.0,800])
plt.tight_layout()
#plt.savefig('killingley.png')
plt.show()
```
